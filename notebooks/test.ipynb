# Latex to Chunks

```python
from kg_agents.ingestion.chunker import chunk_latex_document
from pathlib import Path
```

```python
def main():

    # 🔹 Change this path to your test LaTeX file
    base_dir = Path("../data/latex_outputs")
    latex_file = base_dir / "A First Course in Numerical Methods – Ascher & Greif" / \
                 "29a0ccd7-d6d9-42e9-92c6-ce3bf64facdf" / \
                 "29a0ccd7-d6d9-42e9-92c6-ce3bf64facdf.tex"

    # 🔹 Where chunk JSON will be saved
    output_dir = "../data/chunks"

    print("\n--- Starting Chunking Process ---\n")

    chunks = chunk_latex_document(
        latex_path=latex_file,
        output_dir=output_dir,
        max_chars=4000  # You can adjust this
    )

    print("\n--- Chunking Complete ---")
    print(f"Total chunks created: {len(chunks)}\n")

    # 🔹 Print first 3 chunks for inspection
    for chunk in chunks[:10]:
        print("Chunk ID:", chunk["chunk_id"])
        print("Heading:", chunk["heading"])
        print("Content preview:", chunk["content"][:200])
        print("-" * 50)


if __name__ == "__main__":
    main()
```
